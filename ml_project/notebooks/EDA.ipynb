```python
def dependent_generating(data:pd.DataFrame):
    output = pd.DataFrame()
    indices = data.index
    for feature in data.columns:
        value_counts = data.loc[indices][feature].value_counts(normalize=True)
        keys, values = value_counts.index, value_counts.values
        generated_value = np.random.choice(keys, p=values, size=100)
        indices = set(indices).intersection(data[data[feature] in generated_value].index)
        output[feature] = generated_value
    return output


def independent_generating(data:pd.DataFrame):
    output = pd.DataFrame()
    for feature in data.columns:
        value_counts = data[feature].value_counts(normalize=True)
        keys, values = value_counts.index, value_counts.values
        output[feature] = np.random.choice(keys, p=values, size=100)
    return output


def markov_generating(data:pd.DataFrame):
    output = pd.DataFrame()
    indices = data.index
    for feature in data.columns:
        value_counts = data.loc[indices][feature].value_counts(normalize=True)
        keys, values = value_counts.index, value_counts.values
        generated_value = np.random.choice(keys, p=values)
        indices = data[data[feature] == generated_value].index
        output[feature] = [generated_value]
    return output
```

```python
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
def build_numerical_pipeline() -> Pipeline:
    num_pipeline = Pipeline(
        [
            ("impute", SimpleImputer(missing_values=np.nan, strategy="mean")),
            ("scale", StandardScaler()),
        ]
    )
    return num_pipeline


def make_features(transformer: ColumnTransformer, df: pd.DataFrame) -> pd.DataFrame:
    return pd.DataFrame(transformer.transform(df))


def build_categorical_pipeline() -> Pipeline:
    categorical_pipeline = Pipeline(
        [
            ("impute", SimpleImputer(missing_values=np.nan, strategy="most_frequent")),
            ("ohe", OneHotEncoder()),
        ]
    )
    return categorical_pipeline


def build_transformer(params) -> ColumnTransformer:
    transformer = ColumnTransformer(
        [
            (
                "categorical_pipeline",
                build_categorical_pipeline(),
                ['sex'],
            ),
            (
                "numerical_pipeline",
                build_numerical_pipeline(),
                ['age'],
            ),
        ]
    )
    return transformer
```
